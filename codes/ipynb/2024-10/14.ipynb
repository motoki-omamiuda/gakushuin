```julia
using QuadGK
using Plots

# b = 6
b = 3 * sqrt(3) + 0.001


function P(b)
    var = b ^ 2 * sqrt( Complex(1 - (b ^ 2 / 27)) )
    return (
        (- b ^ 2 + var) ^ (1 / 3) + (- b ^ 2 - var) ^ (1 / 3)
    )
end


function phi(x)
    return 1 / sqrt( 2*x^3 - x^2 + 1/b^2 )
end


plt = plot(
    xlim=(-20, 20), ylim=(-20, 20),
    legend=false,
    ratio=1, # 
    dpi=800, # 
    grid=false,
)


r_list = []
phi_list = []

x_list = []
y_list = []

for i in 0.001:0.001:real(1/P(b))
    push!(r_list, 1/i)
    result, err = quadgk(phi, 0, i)
    push!(phi_list, result)
    print(1/i, "\t", result, "\n")
end

for (phi_val, r_val) in zip(phi_list, r_list)
    push!(x_list, r_val * cos(phi_val))
    push!(y_list, r_val * sin(phi_val) )
end

plot!(plt, x_list, y_list, color=:black, linewidth=2)


phi_end = phi_list[end]

r_list = []
phi_list = []

x_list = []
y_list = []


for i in 0.001:0.001:real(1/P(b))
    push!(r_list, 1/i)
    result, err = quadgk(phi, 0, i)
    push!(phi_list, 2*phi_end - result)
    print(1/i, "\t", result, "\n")
end


for (phi_val, r_val) in zip(phi_list, r_list)
    push!(x_list, r_val * cos(phi_val))
    push!(y_list, r_val * sin(phi_val) )
end

plot!(plt, x_list, y_list, color=:black, linewidth=2)
savefig(plt, "./>3*sqrt(3).png")
plt
```
